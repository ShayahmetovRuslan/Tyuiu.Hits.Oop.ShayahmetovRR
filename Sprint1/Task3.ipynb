### Вариант задания №19


<h2 style="color:DodgerBlue">Описание проекта:</h2>

----

Создать базовый класс Subscription в C#, который будет представлять подписки на
различные услуги. На основе этого класса разработать 2-3 производных класса,
демонстрирующих принципы наследования и полиморфизма. В каждом из классов
должны быть реализованы новые атрибуты и методы, а также переопределены
некоторые методы базового класса для демонстрации полиморфизма.

Требования к базовому классу Subscription:

• Атрибуты: ID подписки (SubscriptionId), Название услуги (ServiceName),
Стоимость подписки (Cost).
    
• Методы:

o CalculateMonthlyCost(): метод для расчета ежемесячной стоимости
подписки.

o ExtendSubscription(): метод для продления подписки на
дополнительный период.

o GetSubscriptionDetails(): метод для получения деталей подписки

Требования к производным классам:

1. ПодпискаНаОнлайнСервис (OnlineServiceSubscription):
Должна содержать
дополнительные атрибуты, такие как Количество доступных пользователей
(MaxUsers). 
Метод CalculateMonthlyCost() должен быть переопределен для
учета количества пользователей при расчете стоимости.

2. ПодпискаНаСтreamинг (StreamingSubscription): 
Должна содержать
дополнительные атрибуты, такие как Количество одновременных потоков
(MaxStreams).
Метод ExtendSubscription() должен быть переопределен для
добавления специальных предложений для продления подписки.

3. ПодпискаНаВидео(VideoSubscription) (если требуется третий класс): 
Должна
содержать дополнительные атрибуты, такие как Качество видео
(VideoQuality). 
Метод GetSubscriptionDetails() должен быть переопределен
для отображения качества видео вместе с другими деталями подписки.

#### Дополнительное задание
Добавьте к сущестующим классам конструктора классов с использованием гетторов и сетторов и реализуйте взаимодействие объектов между собой

<h2 style="color:DodgerBlue">Реализация:</h2>

```
public class Subscription
{
    private int _subscriptionId;
    private string _serviceName;
    private int _cost;
    private int _time;

    public int SubscriptionId
    {
        get {return _subscriptionId;}
        set {_subscriptionId = value;}
    }

    public string ServiceName
    {
        get {return _serviceName;}
        set {_serviceName = value;}
    }

    public int Cost
    {
        get {return _cost;}
        set {_cost = value;}
    }

    public int Time
    {
        get {return _time;}
        set {_time = value;}
    }

    public Subscription() : base()
    {
        Time = 2;
        Cost = 100;
        SubscriptionId = 0;
        ServiceName = "Подписка";
    }

    public virtual void CalculateMonthlyCost()
    {
        Console.WriteLine($"Цена стоимости ежемесячной подписки равна {Cost}.");
    }

    public virtual void ExtendSubscription(int time)
    {
        Time += time;
        Console.WriteLine($"Подписка продлена на {time} месяц(а/ев).");
    }

    public virtual void GetSubscriptionDetails()
    {
        Console.WriteLine($"Цена за месяц: {Cost}\n Оставшееся время: {Time} месяца\n Название подписки: {ServiceName}");
    }
}

    public class OnlineServiceSubscription : Subscription 
    {
        private int _maxUsers = 20;
        private int _currentUsers;

        public OnlineServiceSubscription() : base()
        {
            SubscriptionId = 1;
            ServiceName = "Подписка на онлайн-сервис";
            Cost = 200;
            CurrentUsers = 15;
        }

        public int CurrentUsers
        {
            get {return _currentUsers;}
            set
            {
                if (_currentUsers <= _maxUsers)
                {
                    _currentUsers = value;
                }
                else
                {
                    Console.WriteLine($"Превышено допустимое количество пользователей: {_maxUsers}");
                }
            }
        }

        public override void CalculateMonthlyCost()
        {
            Console.WriteLine($"ID и название подписки: {ServiceName}, {SubscriptionId}.\nТекущее количество пользователей в тарифе: {CurrentUsers}.\nСтоимость подписки: {Cost}.\nПлата в месяц по тарифу: {Cost*CurrentUsers}");
        }
    }

    public class StreamingSubscription : Subscription
    {
        private int _maxStreams = 12;
        private int _currentStreams;

        public int CurrentStreams
        {
            get {return _currentStreams;}
            set
            {
                if (_currentStreams <= _maxStreams)
                {
                    _currentStreams = value;
                }
                else
                {
                    Console.WriteLine($"Превышено допустимое количество стримов: {_maxStreams}");
                }
            }
        }

        public StreamingSubscription() : base()
        {
            SubscriptionId = 2;
            ServiceName = "Подписка на стримы";
            Cost = 400;
            Time = 6;
        }

        public override void ExtendSubscription(int time)
        {
            if (time <= 3)
            {
                Time += time;
                Console.WriteLine($"ID и название подписки: {ServiceName}, {SubscriptionId}.\nСтоимость подписки: {Cost}");
                Console.WriteLine($"Подписка продлена на {time} месяц(а/ев). Оставшееся время: {Time} месяц(а/ев)");
            }
            else
            {
                Time += time;
                Console.WriteLine($"ID и название подписки: {ServiceName}, {SubscriptionId}.\nСтоимость подписки: {Cost}");
                Console.WriteLine($"Подписка продлена на {time} месяц(а/ев). Оставшееся время: {Time} месяц(а/ев)");
                Console.WriteLine($"Вы продлили подписку более чем на 3 месяца! Поздравляем, стоимость подписки на продленный вами период снижена на 25%.");
            }
        }
    }

    public class VideoSubscription : Subscription
    {
        private string _videoQuality;

        public string VideoQuality
        {
            get {return _videoQuality;}
            set {_videoQuality = value;}
        }

        public VideoSubscription() : base()
        {
            SubscriptionId = 3;
            ServiceName = "Подписка на видео";
            Cost = 500;
            Time = 6;
            VideoQuality = "1080p";
        }

        public override void GetSubscriptionDetails()
        {
            Console.WriteLine($"ID и название подписки: {ServiceName}, {SubscriptionId}.\nСтоимость подписки: {Cost}.\nОставшееся время: {Time} месяц(а/ев).\nКачество видео: {VideoQuality}.");
        }
    }

OnlineServiceSubscription ossub = new OnlineServiceSubscription();
ossub.CalculateMonthlyCost();
Console.WriteLine("------------------------------------------");

StreamingSubscription ssub = new StreamingSubscription();
ssub.ExtendSubscription(4);
Console.WriteLine("------------------------------------------");

VideoSubscription vsub = new VideoSubscription();
vsub.GetSubscriptionDetails();

```
